| Flux | Description | Sources |

**Region d'etude :** Maroc — Latitude [27.5 - 35.5] / Longitude [-7.5 - -2.5]  
**Periode d'analyse :** 2022-01-01 au 2025-12-31

```python
# IMPORTATION DES BIBLIOTHEQUES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# ---- Parametres graphiques globaux ----
# Ces parametres s'appliquent uniformement a tous les graphiques du notebook
plt.rcParams.update({
    'figure.facecolor'  : 'white',
    'axes.facecolor'    : '#F8F8F8',
    'axes.grid'         : True,
    'grid.alpha'        : 0.35,
    'grid.linestyle'    : '--',
    'font.family'       : 'DejaVu Sans',
    'axes.spines.top'   : False,
    'axes.spines.right' : False,
    'axes.titlesize'    : 12,
    'axes.labelsize'    : 10,
    'figure.dpi'        : 120,
})

# Palette de couleurs du projet AVORUBY
VERT_FONCE  = '#1A4731'
VERT_CLAIR  = '#2E6B3E'
ROUGE       = '#C0392B'
ORANGE      = '#E67E22'
BLEU        = '#2980B9'
GRIS        = '#7F8C8D'

print('Configuration effectuee avec succes.')
print('Bibliotheques chargees : pandas, numpy, matplotlib, seaborn')
```

## Section 2 — Chargement et Verification des Donnees

```python
# UPLOAD DES FICHIERS DANS GOOGLE COLAB

from google.colab import files

print('Selectionnez les 4 fichiers CSV du projet...')
uploaded = files.upload()

print('\nFichiers recus :')
for nom in uploaded.keys():
    print(f'  - {nom}')
```

```python
# CHARGEMENT DES DATASETS


# --- Flux Ct : donnees climatiques ---
# Source : NASA POWER (temperature, humidite, rayonnement) + Open-Meteo (precipitations)
# Le VPD a ete calcule a partir de T2M et RH2M via la formule de Tetens
df_ct = pd.read_csv('AVORUBY_Dataset_Ct_Final.csv')
df_ct['Date'] = pd.to_datetime(df_ct['Date'])

# Ajout de colonnes temporelles pour les analyses par periode
df_ct['Annee']  = df_ct['Date'].dt.year
df_ct['Mois']   = df_ct['Date'].dt.month
df_ct['Saison'] = df_ct['Mois'].map({
    12:'Hiver', 1:'Hiver',  2:'Hiver',
     3:'Printemps', 4:'Printemps', 5:'Printemps',
     6:'Ete',    7:'Ete',   8:'Ete',
     9:'Automne',10:'Automne',11:'Automne'
})

# Etiquette de risque VPD basee sur des seuils agronomiques etablis
def classer_vpd(valeur):
    if pd.isna(valeur):
        return 'Inconnu'
    if valeur < 0.5:
        return 'Nul'
    if valeur < 1.5:
        return 'Leger'
    if valeur < 3.0:
        return 'Modere'
    return 'Severe'

df_ct['Classe_VPD'] = df_ct['VPD'].apply(classer_vpd)

# --- Flux Ht : disponibilite en eau ---
# Source : NDWI moyen des 18 points (Sentinel-2 via GEE) + surface des barrages
df_ht = pd.read_csv('AVORUBY_Flux_Ht.csv')
df_ht['Date'] = pd.to_datetime(df_ht['Date'])

# --- NDWI spatial : index par point GPS ---
# Ce fichier contient le NDWI mensuel pour chacun des 18 points de la grille
df_ndwi = pd.read_csv('NDWI_Sentinel2.csv')
df_ndwi['Date'] = pd.to_datetime(df_ndwi['Date'])

# --- Barrages : surface en eau par barrage ---
df_barrages = pd.read_csv('Dataset_Barrages_Multi.csv')
df_barrages['Date'] = pd.to_datetime(df_barrages['Date'])

# ---- Verification rapide ----
print('=' * 55)
print('  VERIFICATION DES DATASETS CHARGES')
print('=' * 55)

for nom, df in [('Ct (Climatique)', df_ct), ('Ht (Ressource)', df_ht),
                ('NDWI Spatial', df_ndwi), ('Barrages', df_barrages)]:
    print(f'\n  {nom}')
    print(f'    Dimensions   : {df.shape[0]} lignes x {df.shape[1]} colonnes')
    print(f'    Periode      : {df.Date.min().date()} -> {df.Date.max().date()}')
    print(f'    Colonnes     : {df.columns.tolist()}')

print(f'\n  Points GPS dans Ct : {df_ct[["LAT","LON"]].drop_duplicates().shape[0]} points')
print(f'  Barrages suivis    : {df_barrages["Nom_Barrage"].unique().tolist()}')
print('=' * 55)
```

## Section 3 — Analyse du Flux Ct (Variables Climatiques)

Le Flux Ct contient les variables climatiques journalieres qui determinent le **risque environnemental** pesant sur les cultures :

| Variable | Description | Role |
|----------|-------------|------|
| VPD | Deficit de Pression de Vapeur (kPa) | Stress thermique principal |
| T2M | Temperature a 2m du sol (degC) | Facteur de calcul du VPD |
| RH2M | Humidite relative (%) | Facteur de calcul du VPD |
| Precipitations_mm | Pluie quotidienne (mm) | Apport en eau naturel |
| ALLSKY_SFC_SW_DWN | Rayonnement solaire (MJ/m2/j) | Energie pour la photosynthese |
| GWETTOP | Humidite du sol en surface | Disponibilite d'eau racinaire |

```python
# STATISTIQUES DESCRIPTIVES DU FLUX Ct

variables_ct = ['VPD', 'T2M', 'RH2M', 'Precipitations_mm', 'ALLSKY_SFC_SW_DWN', 'GWETTOP']

print('STATISTIQUES DESCRIPTIVES — Flux Environnemental Ct')
print('Periode : 2022-2025 | Region : Maroc [LAT 27.5-35.5, LON -7.5/-2.5]')
print()
print(df_ct[variables_ct].describe().round(3).to_string())

# Analyse des valeurs manquantes
print('\n--- Valeurs Manquantes ---')
for col in variables_ct:
    nb_nans = df_ct[col].isna().sum()
    pct     = nb_nans / len(df_ct) * 100
    print(f'  {col:<25} : {nb_nans:>5} valeurs manquantes ({pct:.1f}%)')

# Repartition des classes de risque VPD
print('\n--- Repartition des Niveaux de Stress Thermique (VPD) ---')
total_vpd = df_ct['VPD'].notna().sum()
for classe in ['Nul', 'Leger', 'Modere', 'Severe']:
    nb  = (df_ct['Classe_VPD'] == classe).sum()
    pct = nb / total_vpd * 100
    print(f'  {classe:<10} : {nb:>6} jours ({pct:.1f}%)')
```

```python

# GRAPHIQUE 1 : SERIES TEMPORELLES DU FLUX Ct
# Ce graphique montre l'evolution des 4 variables principales
# sur toute la periode d'etude (2022-2025).
# On travaille sur la moyenne de tous les 18 points GPS par jour.

# Agregation : moyenne journaliere sur les 18 points GPS
ct_daily = df_ct.groupby('Date').agg(
    VPD   = ('VPD',               'mean'),
    T2M   = ('T2M',               'mean'),
    Pluie = ('Precipitations_mm', 'mean'),
    RH2M  = ('RH2M',              'mean')
).reset_index()

fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)
fig.suptitle(
    'Flux Environnemental Ct — Evolution Temporelle des Variables Climatiques\n'
    'Moyenne sur 18 points GPS | Region Maroc | 2022-2025',
    fontsize=13, fontweight='bold', y=1.01
)

# --- Sous-graphique 1 : VPD avec zones de risque colorees ---
# Les plages colorees permettent de lire immediatement le niveau de risque
ax = axes[0]
ax.fill_between(ct_daily['Date'], 0,   0.5,  alpha=0.12, color='green',  label='Nul (<0.5)')
ax.fill_between(ct_daily['Date'], 0.5, 1.5,  alpha=0.12, color='gold',   label='Leger (0.5-1.5)')
ax.fill_between(ct_daily['Date'], 1.5, 3.0,  alpha=0.12, color='orange', label='Modere (1.5-3.0)')
ax.fill_between(ct_daily['Date'], 3.0, 9.0,  alpha=0.10, color='red',    label='Severe (>3.0)')
ax.plot(ct_daily['Date'], ct_daily['VPD'], color=VERT_FONCE, linewidth=0.7, alpha=0.8)
# Moyenne mobile 30 jours pour visualiser la tendance sans le bruit journalier
ax.plot(ct_daily['Date'], ct_daily['VPD'].rolling(30, center=True).mean(),
        color=ROUGE, linewidth=2.0, label='Moy. mobile 30 jours')
ax.set_ylabel('VPD (kPa)', fontweight='bold')
ax.set_ylim(0)
ax.legend(loc='upper left', fontsize=8, ncol=5)
ax.set_title('Stress Thermique — VPD (Vapor Pressure Deficit)', fontweight='bold')

# --- Sous-graphique 2 : Temperature ---
ax = axes[1]
ax.plot(ct_daily['Date'], ct_daily['T2M'], color=ROUGE, linewidth=0.7)
ax.fill_between(ct_daily['Date'], ct_daily['T2M'].min(), ct_daily['T2M'],
                alpha=0.12, color=ROUGE)
ax.set_ylabel('Temperature (degC)', fontweight='bold')
ax.set_title('Temperature a 2m du Sol', fontweight='bold')

# --- Sous-graphique 3 : Humidite relative ---
ax = axes[2]
ax.plot(ct_daily['Date'], ct_daily['RH2M'], color=BLEU, linewidth=0.7)
ax.fill_between(ct_daily['Date'], 0, ct_daily['RH2M'], alpha=0.12, color=BLEU)
ax.set_ylabel('Humidite Rel. (%)', fontweight='bold')
ax.set_title('Humidite Relative a 2m', fontweight='bold')

# --- Sous-graphique 4 : Precipitations ---
ax = axes[3]
ax.bar(ct_daily['Date'], ct_daily['Pluie'], color=VERT_CLAIR, alpha=0.75, width=1)
ax.set_ylabel('Pluie (mm/jour)', fontweight='bold')
ax.set_xlabel('Date', fontweight='bold')
ax.set_title('Precipitations Journalieres', fontweight='bold')

plt.tight_layout()
plt.savefig('graphique_01_series_ct.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 1 sauvegarde : graphique_01_series_ct.png')
```

```python
# GRAPHIQUE 2 : HEATMAPS MENSUELLES DU FLUX Ct
# Une heatmap permet de voir d'un coup d'oeil quels mois et
# quelles annees presentent les valeurs les plus elevees.
# Chaque cellule = moyenne mensuelle sur les 18 points GPS.

noms_mois = ['Jan','Fev','Mar','Avr','Mai','Jun',
             'Jul','Aou','Sep','Oct','Nov','Dec']

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
fig.suptitle(
    'Flux Ct — Variation Mensuelle du Stress Thermique et des Precipitations\n'
    'Chaque cellule represente la moyenne mensuelle sur les 18 points GPS',
    fontsize=12, fontweight='bold'
)

# Heatmap 1 : VPD moyen par mois et par annee
pivot_vpd = df_ct.groupby(['Annee','Mois'])['VPD'].mean().unstack(level=1).round(3)
pivot_vpd.columns = noms_mois

sns.heatmap(
    pivot_vpd, annot=True, fmt='.2f',
    cmap='RdYlGn_r',          # Vert = faible stress, Rouge = stress eleve
    vmin=0, vmax=4.5,
    linewidths=0.5, linecolor='white',
    ax=axes[0],
    annot_kws={'size': 9},
    cbar_kws={'label': 'VPD moyen (kPa)'}
)
axes[0].set_title('Stress Thermique — VPD Moyen (kPa)\nVert=Faible | Rouge=Eleve',
                  fontweight='bold')
axes[0].set_xlabel('Mois'); axes[0].set_ylabel('Annee')

# Heatmap 2 : Precipitations moyennes par mois et par annee
pivot_pluie = df_ct.groupby(['Annee','Mois'])['Precipitations_mm'].mean().unstack(level=1).round(2)
pivot_pluie.columns = noms_mois

sns.heatmap(
    pivot_pluie, annot=True, fmt='.2f',
    cmap='Blues',             # Blanc = sec, Bleu fonce = pluie abondante
    linewidths=0.5, linecolor='white',
    ax=axes[1],
    annot_kws={'size': 9},
    cbar_kws={'label': 'Precipitations moy. (mm/j)'}
)
axes[1].set_title('Precipitations Moyennes (mm/jour)\nBlanc=Sec | Bleu=Pluie Abondante',
                  fontweight='bold')
axes[1].set_xlabel('Mois'); axes[1].set_ylabel('Annee')

plt.tight_layout()
plt.savefig('graphique_02_heatmaps_ct.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 2 sauvegarde : graphique_02_heatmaps_ct.png')
```

```python
# GRAPHIQUE 3 : ANALYSE SPATIALE — VPD PAR POINT GPS
# Ce graphique montre les differences de stress thermique
# entre les 18 points de la grille geographique.
# Les points au sud (latitude plus basse) sont generalement
# plus chauds et donc plus stresses.

# Calcul des statistiques par point geographique sur toute la periode
df_spatial = df_ct.groupby(['LAT', 'LON']).agg(
    VPD_moyen     = ('VPD',               'mean'),
    T2M_moyen     = ('T2M',               'mean'),
    Pluie_totale  = ('Precipitations_mm', 'sum'),
    Jours_severe  = ('VPD', lambda x: (x > 3.0).sum())
).reset_index().round(3)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle(
    'Flux Ct — Analyse Spatiale sur la Grille de 18 Points GPS\n'
    'Maroc | 2022-2025',
    fontsize=12, fontweight='bold'
)

# Fond de carte simplifie
for ax in axes:
    ax.set_facecolor('#D6EAF8')  # couleur ocean
    ax.add_patch(plt.Rectangle((-8.5, 27.0), 7.0, 10.0,
                               facecolor='#EAECEE', edgecolor='#AAB7B8',
                               linewidth=0.5, zorder=0))
    ax.set_xlim(-8.2, -2.0)
    ax.set_ylim(27.0, 36.5)
    ax.set_xlabel('Longitude', fontweight='bold')
    ax.set_ylabel('Latitude', fontweight='bold')

# Carte 1 : VPD moyen par point
sc1 = axes[0].scatter(
    df_spatial['LON'], df_spatial['LAT'],
    c=df_spatial['VPD_moyen'], cmap='RdYlGn_r',
    vmin=1.0, vmax=3.0,
    s=df_spatial['VPD_moyen'] * 300,
    alpha=0.85, edgecolors='white', linewidths=1.5, zorder=5
)
plt.colorbar(sc1, ax=axes[0], label='VPD moyen (kPa)', shrink=0.85)
for _, r in df_spatial.iterrows():
    axes[0].annotate(f"{r['VPD_moyen']:.2f}",
        (r['LON'], r['LAT']), xytext=(0, 9), textcoords='offset points',
        ha='center', fontsize=7, fontweight='bold')
axes[0].set_title('VPD Moyen (kPa)', fontweight='bold')

# Carte 2 : Jours de stress severe par point
sc2 = axes[1].scatter(
    df_spatial['LON'], df_spatial['LAT'],
    c=df_spatial['Jours_severe'], cmap='Reds',
    s=df_spatial['Jours_severe'] * 1.5 + 80,
    alpha=0.85, edgecolors='white', linewidths=1.5, zorder=5
)
plt.colorbar(sc2, ax=axes[1], label='Nombre de jours', shrink=0.85)
for _, r in df_spatial.iterrows():
    axes[1].annotate(f"{int(r['Jours_severe'])}j",
        (r['LON'], r['LAT']), xytext=(0, 9), textcoords='offset points',
        ha='center', fontsize=7, fontweight='bold')
axes[1].set_title('Jours de Stress Severe (VPD > 3 kPa)', fontweight='bold')

# Carte 3 : Precipitations totales par point
sc3 = axes[2].scatter(
    df_spatial['LON'], df_spatial['LAT'],
    c=df_spatial['Pluie_totale'], cmap='Blues',
    s=200, alpha=0.85, edgecolors='white', linewidths=1.5, zorder=5
)
plt.colorbar(sc3, ax=axes[2], label='Pluie totale (mm)', shrink=0.85)
for _, r in df_spatial.iterrows():
    axes[2].annotate(f"{int(r['Pluie_totale'])}",
        (r['LON'], r['LAT']), xytext=(0, 9), textcoords='offset points',
        ha='center', fontsize=7, fontweight='bold')
axes[2].set_title('Precipitations Totales (mm) 2022-2025', fontweight='bold')

plt.tight_layout()
plt.savefig('graphique_03_spatial_ct.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 3 sauvegarde : graphique_03_spatial_ct.png')
```

```python
# GRAPHIQUE 4 : DISTRIBUTIONS ET VARIABILITE DU VPD
# Les boxplots permettent de comparer la distribution du VPD
# selon les saisons et les annees, et d'identifier les periodes
# les plus risquees pour les cultures.

df_vpd = df_ct.dropna(subset=['VPD'])

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle(
    'Flux Ct — Variabilite du Stress Thermique (VPD)\n'
    'Par Saison et par Annee | 2022-2025',
    fontsize=12, fontweight='bold'
)

# Boxplot 1 : VPD par saison
# L'ordre est fixe pour suivre le cycle de l'annee
ordre_saisons = ['Hiver', 'Printemps', 'Ete', 'Automne']
couleurs_saisons = {'Hiver': BLEU, 'Printemps': VERT_CLAIR, 'Ete': ROUGE, 'Automne': ORANGE}

data_saisons = [df_vpd[df_vpd['Saison'] == s]['VPD'].values for s in ordre_saisons]
bp = axes[0].boxplot(
    data_saisons,
    labels=ordre_saisons,
    patch_artist=True,
    medianprops={'color': 'white', 'linewidth': 2.5},
    flierprops={'marker': 'o', 'markersize': 2, 'alpha': 0.4}
)
for patch, saison in zip(bp['boxes'], ordre_saisons):
    patch.set_facecolor(couleurs_saisons[saison])
    patch.set_alpha(0.75)

# Ligne de seuil critique
axes[0].axhline(y=3.0, color=ROUGE, linestyle='--', linewidth=1.5,
                alpha=0.7, label='Seuil stress severe (3.0 kPa)')
axes[0].set_ylabel('VPD (kPa)', fontweight='bold')
axes[0].set_title('Distribution par Saison', fontweight='bold')
axes[0].legend(fontsize=8)

# Boxplot 2 : VPD par annee
annees = sorted(df_vpd['Annee'].unique())
data_annees = [df_vpd[df_vpd['Annee'] == a]['VPD'].values for a in annees]
bp2 = axes[1].boxplot(
    data_annees,
    labels=annees,
    patch_artist=True,
    medianprops={'color': 'white', 'linewidth': 2.5},
    flierprops={'marker': 'o', 'markersize': 2, 'alpha': 0.4}
)
for patch in bp2['boxes']:
    patch.set_facecolor(VERT_FONCE)
    patch.set_alpha(0.70)

axes[1].axhline(y=3.0, color=ROUGE, linestyle='--', linewidth=1.5,
                alpha=0.7, label='Seuil stress severe (3.0 kPa)')
axes[1].set_ylabel('VPD (kPa)', fontweight='bold')
axes[1].set_title('Distribution par Annee', fontweight='bold')
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.savefig('graphique_04_boxplots_vpd.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 4 sauvegarde : graphique_04_boxplots_vpd.png')
```

## Section 4 — Analyse du Flux Ht (Ressources en Eau)

Le Flux Ht mesure la **disponibilite effective en eau** pour les cultures, independamment des precipitations.  
Il est constitue de deux composantes complementaires :

| Composante | Methode | Echelle | Interpretation |
|---|---|---|---|
| NDWI | Sentinel-2 via Google Earth Engine | 10-20 metres | Teneur en eau dans la vegetation (stress hydrique interne) |
| Surface Barrages | Donnees officielles (barrages Maroc) | Regional | Reserves strategiques en eau disponibles a l'irrigation |

**Formule NDWI (Gao, 1996) :**  
`NDWI = (B8 - B11) / (B8 + B11)`  
Valeurs : de -1 (vegetation seche) a +1 (vegetation tres bien hydratee)

```python
# GRAPHIQUE 5 : SERIES TEMPORELLES DU FLUX Ht
# Ce graphique montre l'evolution du NDWI et de la surface
# des barrages sur la periode 2022-2025.
# Le NDWI est la moyenne des 18 points GPS, interpolee
# quotidiennement (revisit satellite Sentinel-2 = 5 jours).

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)
fig.suptitle(
    'Flux de Ressource Ht — Evolution Temporelle\n'
    'NDWI (stress hydrique vegetation) et Surface des Barrages | 2022-2025',
    fontsize=13, fontweight='bold', y=1.01
)

# --- NDWI ---
ax = axes[0]

# Zone de reference : NDWI = 0 est la limite theorique stress / hydratation
ax.axhline(y=0, color='black', linewidth=0.8, linestyle='--', alpha=0.5,
           label='Limite hydratation / stress (NDWI = 0)')

# Zone coloree : en dessous de 0 = stress hydrique
ax.fill_between(df_ht['Date'], df_ht['NDWI'], 0,
                where=df_ht['NDWI'] < 0,
                alpha=0.2, color=ROUGE, label='Zone de stress hydrique (NDWI < 0)')
ax.fill_between(df_ht['Date'], df_ht['NDWI'], 0,
                where=df_ht['NDWI'] >= 0,
                alpha=0.2, color=VERT_CLAIR, label='Zone hydratee (NDWI >= 0)')

ax.plot(df_ht['Date'], df_ht['NDWI'], color=VERT_FONCE, linewidth=1.2)
ax.set_ylabel('NDWI moyen', fontweight='bold')
ax.set_title('Index NDWI — Teneur en Eau de la Vegetation\n'
             '(Source : Sentinel-2 via Google Earth Engine, interpole quotidiennement)',
             fontweight='bold', fontsize=10)
ax.legend(fontsize=8, loc='lower right')

# --- Surface des Barrages ---
ax = axes[1]

# Conversion en km2 pour plus de lisibilite (1 m2 = 1e-6 km2)
surface_km2 = df_ht['Surface'] / 1e6

ax.fill_between(df_ht['Date'], 0, surface_km2, alpha=0.3, color=BLEU)
ax.plot(df_ht['Date'], surface_km2, color=BLEU, linewidth=1.2)
ax.set_ylabel('Surface totale (km2)', fontweight='bold')
ax.set_xlabel('Date', fontweight='bold')
ax.set_title('Surface en Eau des Barrages Suivis\n'
             '(Somme des 4 barrages : Oued El Makhazine, Al Wahda, Youssef Ibn Tachfine, Idriss 1er)',
             fontweight='bold', fontsize=10)

plt.tight_layout()
plt.savefig('graphique_05_series_ht.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 5 sauvegarde : graphique_05_series_ht.png')
```

```python
# GRAPHIQUE 6 : COMPARAISON DES 4 BARRAGES
# Ce graphique compare l'evolution de la surface en eau
# pour chacun des 4 barrages suivis dans le projet.
# Chaque barrage couvre une sous-region differente du Maroc.

barrages_liste = df_barrages['Nom_Barrage'].unique()
nb_barrages    = len(barrages_liste)

# Palette distincte pour les 4 barrages
couleurs_barrages = [VERT_FONCE, BLEU, ORANGE, ROUGE]

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle(
    'Flux Ht — Evolution de la Surface en Eau par Barrage\n'
    'Maroc | 2020-2025',
    fontsize=12, fontweight='bold'
)

for ax, barrage, couleur in zip(axes.flat, barrages_liste, couleurs_barrages):
    # Extraction des donnees du barrage en question
    df_b = df_barrages[df_barrages['Nom_Barrage'] == barrage].copy()
    df_b = df_b.sort_values('Date')

    surface_km2 = df_b['Surface'] / 1e6  # conversion m2 -> km2

    ax.fill_between(df_b['Date'], 0, surface_km2, alpha=0.25, color=couleur)
    ax.plot(df_b['Date'], surface_km2, color=couleur, linewidth=1.5)

    # Annotation de la valeur maximale
    idx_max = surface_km2.idxmax()
    ax.annotate(
        f'Max : {surface_km2[idx_max]:.1f} km2',
        xy=(df_b.loc[idx_max, 'Date'], surface_km2[idx_max]),
        xytext=(10, -15), textcoords='offset points',
        fontsize=8, color=couleur, fontweight='bold'
    )

    nom_affiche = barrage.replace('_', ' ')
    ax.set_title(f'Barrage : {nom_affiche}', fontweight='bold')
    ax.set_ylabel('Surface (km2)', fontweight='bold')
    ax.set_xlabel('Date')

plt.tight_layout()
plt.savefig('graphique_06_barrages.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 6 sauvegarde : graphique_06_barrages.png')
```

```python
# GRAPHIQUE 7 : NDWI SPATIAL — CARTE PAR POINT GPS
# Ce graphique utilise le fichier NDWI_Sentinel2.csv qui contient
# le NDWI mensuel pour chacun des 18 points de la grille.
# On compare le NDWI moyen ete vs hiver pour montrer
# la variabilite saisonniere du stress hydrique.

df_ndwi['Mois'] = df_ndwi['Date'].dt.month

# Filtrage : ete (juin-aout) vs hiver (dec-fev) pour le contraste maximum
ndwi_ete    = df_ndwi[df_ndwi['Mois'].isin([6, 7, 8])]
ndwi_hiver  = df_ndwi[df_ndwi['Mois'].isin([12, 1, 2])]

ndwi_ete_geo   = ndwi_ete.groupby(['LAT', 'LON'])['NDWI'].mean().reset_index()
ndwi_hiver_geo = ndwi_hiver.groupby(['LAT', 'LON'])['NDWI'].mean().reset_index()

# Valeurs extremes communes pour la colorbar
vmin = min(ndwi_ete_geo['NDWI'].min(), ndwi_hiver_geo['NDWI'].min())
vmax = max(ndwi_ete_geo['NDWI'].max(), ndwi_hiver_geo['NDWI'].max())

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle(
    'Flux Ht — NDWI Moyen par Point GPS\n'
    'Comparaison Ete (Juin-Aout) vs Hiver (Dec-Fev)',
    fontsize=12, fontweight='bold'
)

for ax, df_geo, titre in [(axes[0], ndwi_ete_geo, 'Ete — Stress Hydrique Maximal'),
                          (axes[1], ndwi_hiver_geo, 'Hiver — Vegetation Mieux Hydratee')]:
    ax.set_facecolor('#D6EAF8')
    ax.add_patch(plt.Rectangle((-8.5, 27.0), 7.0, 10.0,
                               facecolor='#EAECEE', edgecolor='#AAB7B8',
                               linewidth=0.5, zorder=0))

    sc = ax.scatter(
        df_geo['LON'], df_geo['LAT'],
        c=df_geo['NDWI'],
        cmap='RdYlGn',      # Rouge = NDWI bas (stress) / Vert = NDWI haut (hydrate)
        vmin=vmin, vmax=vmax,
        s=200, alpha=0.9,
        edgecolors='white', linewidths=1.5, zorder=5
    )
    plt.colorbar(sc, ax=ax, label='NDWI moyen', shrink=0.85)

    for _, r in df_geo.iterrows():
        ax.annotate(f"{r['NDWI']:.3f}",
            (r['LON'], r['LAT']), xytext=(0, 9), textcoords='offset points',
            ha='center', fontsize=7)

    ax.set_xlim(-8.2, -2.0); ax.set_ylim(27.0, 36.5)
    ax.set_xlabel('Longitude', fontweight='bold')
    ax.set_ylabel('Latitude', fontweight='bold')
    ax.set_title(titre, fontweight='bold')

plt.tight_layout()
plt.savefig('graphique_07_ndwi_spatial.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 7 sauvegarde : graphique_07_ndwi_spatial.png')
```

## Section 5 — Analyse Croisee Ct et Ht : Logique de Compensation

C'est le coeur de la logique AVORUBY CASH.  
Le Flux Ht agit comme un **compensateur** du risque climatique defini par Ct.  

**Principe :**  
- VPD eleve (Ct elevee = risque fort) + NDWI eleve (Ht elevee = eau disponible) **=> Risque compense => Credit possible**  
- VPD eleve (Ct elevee = risque fort) + NDWI bas (Ht faible = pas d'eau) **=> Risque double => Credit refuse**

**Formule du Score SABC :**  
`SABC = sigma( Production_Predite x Prix_Estime / Risque_Brut x (1 - Gamma) x (1 - Ht) )`

```python
# GRAPHIQUE 8 : ANALYSE CROISEE Ct vs Ht
# Ce graphique superpose VPD et NDWI sur le meme axe temporel
# pour montrer leur relation inverse :
# quand le VPD monte (stress thermique), le NDWI baisse (stress hydrique).
# Les periodes ou les deux indicateurs sont defavorables simultanement
# sont les plus risquees pour accorder un credit.

# Agregation mensuelle pour un graphique lisible
ct_mensuel = df_ct.groupby(df_ct['Date'].dt.to_period('M'))['VPD'].mean().reset_index()
ct_mensuel['Date'] = ct_mensuel['Date'].dt.to_timestamp()

ht_mensuel = df_ht.groupby(df_ht['Date'].dt.to_period('M'))[['NDWI','Surface']].mean().reset_index()
ht_mensuel['Date'] = ht_mensuel['Date'].dt.to_timestamp()

# Fusion des deux flux sur l'axe temporel commun
df_croise = pd.merge(ct_mensuel, ht_mensuel, on='Date', how='inner')

fig, ax1 = plt.subplots(figsize=(14, 6))
ax2 = ax1.twinx()  # deuxieme axe partageant le meme axe des x

# VPD sur l'axe gauche (rouge)
ax1.plot(df_croise['Date'], df_croise['VPD'], color=ROUGE, linewidth=2.5,
         label='VPD mensuel moyen (Ct)')
ax1.fill_between(df_croise['Date'], 0, df_croise['VPD'],
                 alpha=0.08, color=ROUGE)
ax1.axhline(y=3.0, color=ROUGE, linestyle=':', linewidth=1.2, alpha=0.6)
ax1.set_ylabel('VPD (kPa) — Flux Ct', color=ROUGE, fontweight='bold', fontsize=11)
ax1.tick_params(axis='y', labelcolor=ROUGE)
ax1.set_ylim(0)

# NDWI sur l'axe droit (vert)
ax2.plot(df_croise['Date'], df_croise['NDWI'], color=VERT_FONCE, linewidth=2.5,
         linestyle='--', label='NDWI mensuel moyen (Ht)')
ax2.axhline(y=0, color=VERT_FONCE, linestyle=':', linewidth=1.2, alpha=0.6)
ax2.set_ylabel('NDWI moyen — Flux Ht', color=VERT_FONCE, fontweight='bold', fontsize=11)
ax2.tick_params(axis='y', labelcolor=VERT_FONCE)

# Mise en commun des legendes des deux axes
lignes1, labels1 = ax1.get_legend_handles_labels()
lignes2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lignes1 + lignes2, labels1 + labels2, loc='upper left', fontsize=9)

ax1.set_xlabel('Date', fontweight='bold')
ax1.set_title(
    'Relation entre le Stress Thermique (VPD — Flux Ct) et le Stress Hydrique (NDWI — Flux Ht)\n'
    'Une hausse du VPD correspond generalement a une baisse du NDWI => risque cumule',
    fontweight='bold', fontsize=11
)

plt.tight_layout()
plt.savefig('graphique_08_croise_ct_ht.png', dpi=150, bbox_inches='tight')
plt.show()
print('Graphique 8 sauvegarde : graphique_08_croise_ct_ht.png')
```

```python
# GRAPHIQUE 9 : MATRICE DE CORRELATION
# La matrice de correlation mesure la force du lien lineaire
# entre toutes les variables des deux flux.
# Valeur proche de +1 : les deux variables augmentent ensemble.
# Valeur proche de -1 : quand l'une monte, l'autre descend.
# Valeur proche de 0  : pas de relation lineaire.

# Fusion Ct et Ht sur l'axe mensuel pour avoir toutes les variables ensemble
df_corr = pd.merge(ct_mensuel, ht_mensuel, on='Date', how='inner')
df_corr.rename(columns={'VPD': 'VPD (Ct)', 'NDWI': 'NDWI (Ht)',
                        'Surface': 'Surface Barrages (Ht)'}, inplace=True)

# Ajout des variables Ct mensuelles
ct_plus = df_ct.groupby(df_ct['Date'].dt.to_period('M')).agg(
    T2M   = ('T2M',               'mean'),
    RH2M  = ('RH2M',              'mean'),
    Pluie = ('Precipitations_mm', 'mean')
).reset_index()
ct_plus['Date'] = ct_plus['Date'].dt.to_timestamp()

df_corr = pd.merge(df_corr, ct_plus, on='Date', how='inner')
df_corr.rename(columns={'T2M': 'Temperature (Ct)', 'RH2M': 'Humidite (Ct)',
                        'Pluie': 'Precipitations (Ct)'}, inplace=True)

# Selection des variables pour la matrice
vars_corr = ['VPD (Ct)', 'Temperature (Ct)', 'Humidite (Ct)',
             'Precipitations (Ct)', 'NDWI (Ht)', 'Surface Barrages (Ht)']

matrice = df_corr[vars_corr].corr().round(3)

fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(
    matrice,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn',
    vmin=-1, vmax=1,
    linewidths=0.6,
    linecolor='white',
    square=True,
    ax=ax,
    annot_kws={'size': 10, 'weight': 'bold'},
    cbar_kws={'label': 'Coefficient de Correlation de Pearson', 'shrink': 0.85}
)
ax.set_title(
    'Matrice de Correlation — Variables Ct et Ht\n'
    'Vert = Correlation positive | Rouge = Correlation negative | 0 = Independance',
    fontweight='bold', fontsize=11, pad=15
)

plt.tight_layout()
plt.savefig('graphique_09_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

# Identification des correlations significatives avec le VPD
print('Correlations avec le VPD (par ordre decroissant de valeur absolue) :\n')
vpd_corr = matrice['VPD (Ct)'].drop('VPD (Ct)').sort_values(key=abs, ascending=False)
for var, val in vpd_corr.items():
    sens = 'correlation positive' if val > 0 else 'correlation negative'
    force = 'FORTE' if abs(val) > 0.6 else 'Moderee' if abs(val) > 0.3 else 'Faible'
    print(f'  {var:<30} r = {val:+.3f}  =>  {sens} ({force})')
print('\nGraphique 9 sauvegarde : graphique_09_correlation.png')
```

Ce notebook a presente l'analyse complete des deux premiers flux du projet AVORUBY CASH :

**Flux Ct (Environnemental) :**  
Les donnees NASA POWER et Open-Meteo confirment un stress thermique significatif sur la region d'etude,  
particulierement durant l'ete (VPD superieur a 3 kPa en juillet-aout).

L'analyse croisee montre que Ct et Ht sont inversement correles :

C'est precisement ce que le score SABC est concu a quantifier et a differencier.
