# Notebook 04 — Modélisation XGBoost + LSTM (PyTorch)
**AVORUBY CASH** — Prédiction du risque climatique + Score SABC

**Cible :** VPD à J+7 (risque thermique futur) + Decision_credit
**Modèles :** XGBoost (tabular) + LSTM PyTorch (time series)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os, joblib, warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import xgboost as xgb
import shap
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

BASE_PATH = os.path.dirname(os.path.abspath("__file__"))
PROC_PATH = os.path.join(BASE_PATH, "..", "processed")
MODEL_PATH = os.path.join(BASE_PATH, "..", "models")
os.makedirs(MODEL_PATH, exist_ok=True)

print(" Librairies importées")
print(" PyTorch version:", torch.__version__)
```

## 1. Chargement des données

```python
df = pd.read_csv(os.path.join(PROC_PATH, "flux_Ct_Ht_features.csv"))
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date").reset_index(drop=True)

ag = pd.read_csv(os.path.join(BASE_PATH, "..", "AVR_DATA", "Dataset_Agriculteurs_Complet.csv"))

print(" Features Ct+Ht:", df.shape)
print(" Agriculteurs:", ag.shape)
print(df.head(3))
```

## 2. Préparation — Cible VPD à J+7

```python
df["VPD_J7"] = df["VPD"].shift(-7)
df = df.dropna(subset=["VPD_J7"]).reset_index(drop=True)

EXCLUDE = ["Date", "VPD_J7"]
features = [c for c in df.columns if c not in EXCLUDE]

X = df[features]
y = df["VPD_J7"]

print(f" Features: {len(features)}")
print(f" Samples: {len(X)}")
print(f" VPD_J7 — min: {y.min():.2f}, max: {y.max():.2f}, mean: {y.mean():.2f}")
```

## 3. Split temporel

```python
split_date = "2024-09-30"
train_mask = df["Date"] <= split_date
test_mask  = df["Date"] >  split_date

X_train, X_test = X[train_mask], X[test_mask]
y_train, y_test = y[train_mask], y[test_mask]

print(f" Train: {X_train.shape[0]} jours ({df[train_mask].Date.min().date()} → {df[train_mask].Date.max().date()})")
print(f" Test : {X_test.shape[0]}  jours ({df[test_mask].Date.min().date()} → {df[test_mask].Date.max().date()})")
```

## 4. Modèle 1 — XGBoost

```python
xgb_model = xgb.XGBRegressor(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
    verbosity=0
)

xgb_model.fit(X_train, y_train,
              eval_set=[(X_test, y_test)],
              verbose=False)

y_pred_xgb = xgb_model.predict(X_test)

mae_xgb  = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2_xgb   = r2_score(y_test, y_pred_xgb)

print("=== XGBoost — Résultats ===")
print(f"MAE  : {mae_xgb:.4f} kPa")
print(f"RMSE : {rmse_xgb:.4f} kPa")
print(f"R²   : {r2_xgb:.4f}")
```

```python
dates_test = df[test_mask]["Date"].values

plt.figure(figsize=(14, 5))
plt.plot(dates_test, y_test.values, label="VPD Réel", color="steelblue", linewidth=1)
plt.plot(dates_test, y_pred_xgb,    label="VPD Prédit (XGBoost)", color="tomato", linewidth=1, alpha=0.8)
plt.title(f"XGBoost — Prédiction VPD J+7 | R²={r2_xgb:.3f} | MAE={mae_xgb:.3f} kPa")
plt.ylabel("VPD (kPa)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(PROC_PATH, "xgboost_predictions.png"), dpi=150)
plt.show()
```

```python
explainer   = shap.Explainer(xgb_model)
shap_values = explainer(X_test)

plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_test, max_display=15, show=False)
plt.title("SHAP — Features les plus importantes (XGBoost)")
plt.tight_layout()
plt.savefig(os.path.join(PROC_PATH, "shap_importance.png"), dpi=150, bbox_inches="tight")
plt.show()
print(" SHAP calculé")
```

```python
joblib.dump(xgb_model, os.path.join(MODEL_PATH, "xgboost_vpd.pkl"))
print(" XGBoost sauvegardé: models/xgboost_vpd.pkl")
```

## 5. Modèle 2 — LSTM (PyTorch)

```python
# Normalisation
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_sc = scaler_X.fit_transform(X_train)
X_test_sc  = scaler_X.transform(X_test)
y_train_sc = scaler_y.fit_transform(y_train.values.reshape(-1,1)).ravel()
y_test_sc  = scaler_y.transform(y_test.values.reshape(-1,1)).ravel()

# Fenêtre glissante de 30 jours
WINDOW = 30

def create_sequences(X, y, window):
    Xs, ys = [], []
    for i in range(len(X) - window):
        Xs.append(X[i:i+window])
        ys.append(y[i+window])
    return np.array(Xs, dtype=np.float32), np.array(ys, dtype=np.float32)

X_train_seq, y_train_seq = create_sequences(X_train_sc, y_train_sc, WINDOW)
X_test_seq,  y_test_seq  = create_sequences(X_test_sc,  y_test_sc,  WINDOW)

print(f" Train séquences: {X_train_seq.shape}")
print(f" Test  séquences: {X_test_seq.shape}")
```

```python
# Architecture LSTM PyTorch
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden1=64, hidden2=32):
        super(LSTMModel, self).__init__()
        self.lstm1   = nn.LSTM(input_size, hidden1, batch_first=True)
        self.drop1   = nn.Dropout(0.2)
        self.lstm2   = nn.LSTM(hidden1, hidden2, batch_first=True)
        self.drop2   = nn.Dropout(0.2)
        self.fc      = nn.Linear(hidden2, 1)

    def forward(self, x):
        out, _ = self.lstm1(x)
        out     = self.drop1(out)
        out, _ = self.lstm2(out)
        out     = self.drop2(out[:, -1, :])
        return self.fc(out).squeeze()

INPUT_SIZE = X_train_seq.shape[2]
model_lstm = LSTMModel(INPUT_SIZE)
print(model_lstm)
print(f" Paramètres: {sum(p.numel() for p in model_lstm.parameters()):,}")
```

```python
# DataLoaders
train_dataset = TensorDataset(torch.tensor(X_train_seq), torch.tensor(y_train_seq))
test_dataset  = TensorDataset(torch.tensor(X_test_seq),  torch.tensor(y_test_seq))
train_loader  = DataLoader(train_dataset, batch_size=32, shuffle=False)
test_loader   = DataLoader(test_dataset,  batch_size=32, shuffle=False)

# Entraînement
optimizer  = torch.optim.Adam(model_lstm.parameters(), lr=0.001)
criterion  = nn.MSELoss()
EPOCHS     = 100
patience   = 10
best_loss  = float("inf")
patience_counter = 0
train_losses, val_losses = [], []

for epoch in range(EPOCHS):
    model_lstm.train()
    train_loss = 0
    for Xb, yb in train_loader:
        optimizer.zero_grad()
        pred = model_lstm(Xb)
        loss = criterion(pred, yb)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    train_loss /= len(train_loader)

    model_lstm.eval()
    val_loss = 0
    with torch.no_grad():
        for Xb, yb in test_loader:
            pred = model_lstm(Xb)
            val_loss += criterion(pred, yb).item()
    val_loss /= len(test_loader)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    if val_loss < best_loss:
        best_loss = val_loss
        torch.save(model_lstm.state_dict(), os.path.join(MODEL_PATH, "lstm_vpd_best.pt"))
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping à epoch {epoch+1}")
            break

    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1:3d} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
```

```python
# Courbe de loss
plt.figure(figsize=(10, 4))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses,   label="Val Loss")
plt.title("LSTM PyTorch — Courbe d apprentissage")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(PROC_PATH, "lstm_loss.png"), dpi=150)
plt.show()
```

```python
# Métriques LSTM
model_lstm.load_state_dict(torch.load(os.path.join(MODEL_PATH, "lstm_vpd_best.pt")))
model_lstm.eval()

with torch.no_grad():
    y_pred_lstm_sc = model_lstm(torch.tensor(X_test_seq)).numpy()

y_pred_lstm = scaler_y.inverse_transform(y_pred_lstm_sc.reshape(-1,1)).ravel()
y_test_lstm = scaler_y.inverse_transform(y_test_seq.reshape(-1,1)).ravel()

mae_lstm  = mean_absolute_error(y_test_lstm, y_pred_lstm)
rmse_lstm = np.sqrt(mean_squared_error(y_test_lstm, y_pred_lstm))
r2_lstm   = r2_score(y_test_lstm, y_pred_lstm)

print("=== LSTM PyTorch — Résultats ===")
print(f"MAE  : {mae_lstm:.4f} kPa")
print(f"RMSE : {rmse_lstm:.4f} kPa")
print(f"R²   : {r2_lstm:.4f}")
```

```python
plt.figure(figsize=(14, 5))
plt.plot(y_test_lstm,  label="VPD Réel",          color="steelblue",  linewidth=1)
plt.plot(y_pred_lstm,  label="VPD Prédit (LSTM)", color="darkorange", linewidth=1, alpha=0.8)
plt.title(f"LSTM PyTorch — Prédiction VPD J+7 | R²={r2_lstm:.3f} | MAE={mae_lstm:.3f} kPa")
plt.ylabel("VPD (kPa)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(PROC_PATH, "lstm_predictions.png"), dpi=150)
plt.show()
```

```python
joblib.dump(scaler_X, os.path.join(MODEL_PATH, "scaler_X.pkl"))
joblib.dump(scaler_y, os.path.join(MODEL_PATH, "scaler_y.pkl"))
print("✅ Scalers sauvegardés")
print("✅ LSTM sauvegardé: models/lstm_vpd_best.pt")
```

## 6. Comparaison XGBoost vs LSTM

```python
resultats = pd.DataFrame({
    "Modèle":     ["XGBoost", "LSTM (PyTorch)"],
    "MAE (kPa)":  [mae_xgb,  mae_lstm],
    "RMSE (kPa)": [rmse_xgb, rmse_lstm],
    "R²":         [r2_xgb,   r2_lstm]
}).round(4)

print("=== Tableau comparatif ===")
print(resultats.to_string(index=False))

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
metrics = ["MAE (kPa)", "RMSE (kPa)", "R²"]
colors  = ["steelblue", "darkorange"]

for i, metric in enumerate(metrics):
    axes[i].bar(resultats["Modèle"], resultats[metric], color=colors)
    axes[i].set_title(metric)

fig.suptitle("XGBoost vs LSTM PyTorch — Comparaison", fontweight="bold")
plt.tight_layout()
plt.savefig(os.path.join(PROC_PATH, "comparaison_modeles.png"), dpi=150)
plt.show()
```

## 7. Classification — Décision de crédit (SABC)

```python
le = LabelEncoder()
ag["Decision_encoded"] = le.fit_transform(ag["Decision_credit"])

cat_cols = ["Culture", "Localisation", "Type_Sol"]
ag_enc = pd.get_dummies(ag, columns=cat_cols)

feature_cols = [c for c in ag_enc.columns if c not in
               ["ID_Agriculteur","Decision_credit","Decision_encoded","SABC_score"]]
X_ag = ag_enc[feature_cols]
y_ag = ag_enc["Decision_encoded"]

split_idx = int(len(ag_enc) * 0.8)
X_ag_train, X_ag_test = X_ag.iloc[:split_idx], X_ag.iloc[split_idx:]
y_ag_train, y_ag_test = y_ag.iloc[:split_idx], y_ag.iloc[split_idx:]

print(f"✅ Train: {X_ag_train.shape}")
print(f"✅ Test : {X_ag_test.shape}")
```

```python
clf = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.05,
    random_state=42,
    n_jobs=-1,
    verbosity=0
)

clf.fit(X_ag_train, y_ag_train)
y_pred_clf = clf.predict(X_ag_test)

print("=== Classification — Rapport ===")
print(classification_report(y_ag_test, y_pred_clf, target_names=le.classes_))

cm   = confusion_matrix(y_ag_test, y_pred_clf)
disp = ConfusionMatrixDisplay(cm, display_labels=le.classes_)
disp.plot(cmap="Blues")
plt.title("Matrice de confusion — Décision Crédit")
plt.tight_layout()
plt.savefig(os.path.join(PROC_PATH, "confusion_matrix.png"), dpi=150)
plt.show()
```

```python
joblib.dump(clf, os.path.join(MODEL_PATH, "xgboost_credit.pkl"))
joblib.dump(le,  os.path.join(MODEL_PATH, "label_encoder.pkl"))
print("✅ Classificateur sauvegardé: models/xgboost_credit.pkl")
print()
print("🎉 Modélisation terminée — tous les modèles sont dans models/")
```
