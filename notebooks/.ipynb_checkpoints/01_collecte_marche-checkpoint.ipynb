```python
# Charger les 3 fichiers Avocat Values
avocat_val_files = [
    os.path.join(RAW_PATH, "RawAvocat", f) 
    for f in os.listdir(os.path.join(RAW_PATH, "RawAvocat"))
]

avocat_val = pd.concat([pd.read_excel(f, header=10, engine='openpyxl') for f in avocat_val_files])
print("✅ Avocat Values chargé :", avocat_val.shape)
print(avocat_val.head(3))
```

Maintenant on voit 2 problèmes à régler :

Années 2020/2021 présentes → à exclure
Pays non-EU présents (World, Mauritania, Oman...) → à filtrer

```python
# Liste pays EU
pays_eu = ["Netherlands", "France", "Germany", "Spain", "Poland", 
           "Portugal", "Italy", "Belgium", "Sweden", "Denmark",
           "Austria", "Finland", "Ireland", "Czechia", "Romania"]

def nettoyer_trademap(df):
    # 1. Garder seulement pays EU
    df = df[df["Importers"].isin(pays_eu)].copy()
    
    # 2. Garder seulement colonnes 2022-2025 + Importers
    cols_date = [c for c in df.columns 
                 if any(str(y) in str(c) for y in [2022, 2023, 2024, 2025])]
    df = df[["Importers"] + cols_date]
    
    # 3. Reset index
    df = df.reset_index(drop=True)
    return df

avocat_val_clean = nettoyer_trademap(avocat_val)
print("✅ Après nettoyage:", avocat_val_clean.shape)
print(avocat_val_clean["Importers"].unique())
```

Cellule 3 — Charger les fichiers Avocat (Quantities)

```python
avocat_qty_files = [
    os.path.join(RAW_PATH, "rawAvocatquantity", f) 
    for f in os.listdir(os.path.join(RAW_PATH, "rawAvocatquantity"))
]

avocat_qty = pd.concat([pd.read_excel(f, header=10, engine='openpyxl') for f in avocat_qty_files])
print("✅ Avocat Quantities chargé :", avocat_qty.shape)
print(avocat_qty.head(3))
```

Cellule 4 — Charger les fichiers Ruby (Values + Quantities)

```python
ruby_val_files = [
    os.path.join(RAW_PATH, "RawRuby", f) 
    for f in os.listdir(os.path.join(RAW_PATH, "RawRuby"))
]
ruby_qty_files = [
    os.path.join(RAW_PATH, "rawRubyquantity", f) 
    for f in os.listdir(os.path.join(RAW_PATH, "rawRubyquantity"))
]

ruby_val = pd.concat([pd.read_excel(f, header=10, engine='openpyxl') for f in ruby_val_files])
ruby_qty = pd.concat([pd.read_excel(f, header=10, engine='openpyxl') for f in ruby_qty_files])
print("✅ Ruby Values chargé :", ruby_val.shape)
print("✅ Ruby Quantities chargé :", ruby_qty.shape)
```

Cellule 5 — Charger EUR/MAD

```python
eur_mad = pd.read_excel(os.path.join(RAW_PATH, "Bank_Almaghrib.xlsx"), header=3)
print("✅ EUR/MAD chargé :", eur_mad.shape)
print(eur_mad.head(3))
```

Maintenant applique le même nettoyage aux 3 autres datasets :

```python
avocat_qty_clean = nettoyer_trademap(avocat_qty)
ruby_val_clean   = nettoyer_trademap(ruby_val)
ruby_qty_clean   = nettoyer_trademap(ruby_qty)

print("✅ Avocat Quantities:", avocat_qty_clean.shape)
print("✅ Ruby Values:", ruby_val_clean.shape)
print("✅ Ruby Quantities:", ruby_qty_clean.shape)
```

Cellule suivante — Transformer en format long (date en colonne)
Le tableau Trade Map était en format "large" — une colonne par mois. On l'a transformé en format "long" — une ligne par mois par pays. C'est le format qu'attend pandas et le LSTM.

```python
def wide_to_long(df, value_name):
    df_long = df.melt(
        id_vars="Importers",
        var_name="date",
        value_name=value_name
    )
    # Nettoyer la date : "Exported value in 2022-M01" → "2022-01"
    df_long["date"] = df_long["date"].str.extract(r'(\d{4}-M\d{2})')
    df_long["date"] = df_long["date"].str.replace("-M", "-")
    df_long = df_long.dropna(subset=["date", value_name])
    df_long = df_long.sort_values(["Importers", "date"]).reset_index(drop=True)
    return df_long

avocat_val_long = wide_to_long(avocat_val_clean, "avocat_value_usd")
avocat_qty_long = wide_to_long(avocat_qty_clean, "avocat_qty_kg")
ruby_val_long   = wide_to_long(ruby_val_clean,   "ruby_value_usd")
ruby_qty_long   = wide_to_long(ruby_qty_clean,   "ruby_qty_kg")

print("✅ Avocat Val long:", avocat_val_long.shape)
print(avocat_val_long.head(3))
```

Agréger par date (sommer tous les pays EU)
Au lieu d'avoir une ligne par pays par mois, on veut une seule ligne par mois = total EU.

```python
# Sommer tous les pays EU par mois → 1 valeur EU par mois
avocat_val_eu = avocat_val_long.groupby("date")["avocat_value_usd"].sum().reset_index()
avocat_qty_eu = avocat_qty_long.groupby("date")["avocat_qty_kg"].sum().reset_index()
ruby_val_eu   = ruby_val_long.groupby("date")["ruby_value_usd"].sum().reset_index()
ruby_qty_eu   = ruby_qty_long.groupby("date")["ruby_qty_kg"].sum().reset_index()

print("✅ Avocat Value EU par mois:")
print(avocat_val_eu.head(5))
```

Cellule suivante — Fusionner toutes les séries + EUR/MAD On assemble tout en un seul DataFrame propre

```python
# Charger EUR/MAD
eur_mad = pd.read_excel(
    os.path.join(RAW_PATH, "Bank_Almaghrib.xlsx"), 
    header=3, engine='openpyxl'
)
eur_mad = eur_mad.iloc[:, :2]
eur_mad.columns = ["date", "eur_mad"]
eur_mad = eur_mad.dropna()
print("✅ EUR/MAD:", eur_mad.head(3))

# Fusionner toutes les séries sur la date
from functools import reduce
dfs = [avocat_val_eu, avocat_qty_eu, ruby_val_eu, ruby_qty_eu, eur_mad]
flux_marche = reduce(lambda l, r: pd.merge(l, r, on="date", how="inner"), dfs)

print("\n✅ Flux Marché fusionné:", flux_marche.shape)
print(flux_marche.head(5))
```

Ce qu'on vient de faire — Fusion de toutes les sources :
On a combiné les 5 tableaux séparés en un seul tableau unifié par date.
avocat_val_eu  ─┐
avocat_qty_eu  ─┤
ruby_val_eu    ─┼─→ MERGE par "date" → flux_marche (41 lignes x 6 colonnes)
ruby_qty_eu    ─┤
eur_mad        ─┘

Cellule suivante — Calculer le prix moyen en MAD/kg
On convertit les valeurs USD en MAD et on calcule le prix réel par kg :

```python
# Convertir toutes les colonnes numériques
cols_num = ["avocat_value_usd", "avocat_qty_kg", "ruby_value_usd", "ruby_qty_kg", "eur_mad"]
for col in cols_num:
    flux_marche[col] = pd.to_numeric(flux_marche[col], errors='coerce')

# Maintenant calculer les prix
flux_marche["avocat_prix_mad_kg"] = (
    (flux_marche["avocat_value_usd"] * 1000 / 1.08 * flux_marche["eur_mad"]) 
    / flux_marche["avocat_qty_kg"].where(flux_marche["avocat_qty_kg"] > 0, np.nan)
).round(2)

flux_marche["ruby_prix_mad_kg"] = (
    (flux_marche["ruby_value_usd"] * 1000 / 1.08 * flux_marche["eur_mad"]) 
    / flux_marche["ruby_qty_kg"].where(flux_marche["ruby_qty_kg"] > 0, np.nan)
).round(2)

print("✅ Prix corrigés (MAD/kg):")
print(flux_marche[["date", "avocat_prix_mad_kg", "ruby_prix_mad_kg"]].head(8))
```

ATTENTION A CORRIGER 
Données Trade Map → en USD
Taux BAM téléchargé → EUR/MAD

Les NaN viennent des mois où avocat_qty_kg = 0 (pas d'export ce mois-là) → division impossible 

Cellule suivante — Calculer la volatilité σ des prix
La volatilité = écart-type glissant sur 3 mois — mesure l'instabilité des prix :

```python
# Trier par date
flux_marche = flux_marche.sort_values("date").reset_index(drop=True)

# Volatilité rolling 3 mois
flux_marche["avocat_volatilite"] = (
    flux_marche["avocat_prix_mad_kg"].rolling(window=3, min_periods=2).std().round(2)
)
flux_marche["ruby_volatilite"] = (
    flux_marche["ruby_prix_mad_kg"].rolling(window=3, min_periods=2).std().round(2)
)

print("✅ Volatilité calculée:")
print(flux_marche[["date", "avocat_prix_mad_kg", "avocat_volatilite", 
                    "ruby_prix_mad_kg", "ruby_volatilite"]].head(10))
```

On va combler les NAN par interpolation :

```python
# Interpolation linéaire pour combler les NaN
flux_marche["avocat_prix_mad_kg"] = (
    flux_marche["avocat_prix_mad_kg"].interpolate(method="linear")
)
flux_marche["ruby_prix_mad_kg"] = (
    flux_marche["ruby_prix_mad_kg"].interpolate(method="linear")
)

# Recalculer la volatilité après interpolation
flux_marche["avocat_volatilite"] = (
    flux_marche["avocat_prix_mad_kg"].rolling(window=3, min_periods=2).std().round(2)
)
flux_marche["ruby_volatilite"] = (
    flux_marche["ruby_prix_mad_kg"].rolling(window=3, min_periods=2).std().round(2)
)

# Vérifier les NaN restants
print("NaN restants:")
print(flux_marche.isnull().sum())
print("\n✅ Aperçu:")
print(flux_marche[["date","avocat_prix_mad_kg","ruby_prix_mad_kg"]].head(10))
```

1 NaN restant dans chaque volatilité — c'est la première ligne qui n'a pas assez de valeurs précédentes pour calculer l'écart-type. On le remplace par 0 :

```python
# Remplacer le seul NaN restant dans volatilité par 0
flux_marche["avocat_volatilite"] = flux_marche["avocat_volatilite"].fillna(0)
flux_marche["ruby_volatilite"]   = flux_marche["ruby_volatilite"].fillna(0)

print("✅ Aucun NaN restant:")
print(flux_marche.isnull().sum())
```

 Sauvegarder dans processed/ :

```python
# Sauvegarder le fichier final
output_path = os.path.join(PROC_PATH, "flux_marche.csv")
flux_marche.to_csv(output_path, index=False)

print("✅ Fichier sauvegardé :", output_path)
print("📊 Shape finale :", flux_marche.shape)
print("📋 Colonnes :", flux_marche.columns.tolist())
```
